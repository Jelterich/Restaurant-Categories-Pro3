```python
import json
import pandas as pd
from pathlib import Path

# Make a reference to the comic_books_expanded.csv file path
csv_path = Path("filtered_restaurants2.csv")

# Import the comic_books_expanded.csv file as a DataFrame
restaurant_df = pd.read_csv(csv_path, encoding="utf-8")

# Check the special characters imported correctly
restaurant_df.tail()
```

```python
clean_restaurant_df = restaurant_df.drop(columns=['business_id', 'address','latitude','longitude','attributes','hours'])
clean_restaurant_df
```

```python
unique_cities_states_df = clean_restaurant_df[['city', 'state']].drop_duplicates()
unique_cities_states_df
```

```python
city_counts_df = clean_restaurant_df['city'].value_counts()
sorted_city_counts_df = city_counts_df.sort_values(ascending=False)
sorted_city_counts_df
```

```python
state_counts_df = clean_restaurant_df['state'].value_counts()
sorted_state_counts_df = state_counts_df.sort_values(ascending=False)
sorted_state_counts_df
```
